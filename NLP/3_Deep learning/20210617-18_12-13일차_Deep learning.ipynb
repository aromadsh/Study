```python
import requests
from bs4 import BeautifulSoup as BS
import pandas as pd
import numpy as np
import re
```

```python
def refine_price(text):
    price = int(text.replace(",",""))
    return price
```

```python
df = pd.read_csv('st1.csv', encoding="CP949")

for i in range(1,3):
    stock_code = df["종목코드"][i]
    
    if len(df["종목코드"][i].astype(str)) != 6:
        df["종목코드"][i] = '{:06d}'.format(df['종목코드'][i])        

    url = f"https://finance.naver.com/item/sise_day.nhn?code={stock_code}"
    res = requests.get(url)
    soup = BS(res.text, "html.parser")

stock = []

for i in range(1, 6):
    url = f"https://finance.naver.com/item/sise_day.nhn?code={stock_code}&page={i}"
    headers = {"user-agent" : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36'}
    res = requests.get(url, headers = headers)
    soup = BS(res.text, 'html.parser')
    result = soup.find_all('tr',{"onmouseover":"mouseOver(this)"})
    
    for tr in result:
        date = tr.find_all("td")[0].text
        temp = []
        for td in tr.find_all("td")[1:]:
            price = refine_price(td.text)
            temp.append(price)
        stock.append([date]+temp)
col = ["날짜", "종가", "전일비", "시가", "고가", "저가", "거래량"]
stock_df = pd.DataFrame(stock, columns = col)


```
