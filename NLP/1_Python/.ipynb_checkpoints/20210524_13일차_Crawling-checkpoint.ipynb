### 실습 02

- 사이트내에 인기검색종목과 주요해외지수를 각각 크롤링하여 종목명과 주가지수를 아래와 같이 리스트로 정리해주세요.

```python
url = "https://scrapying-study.firebaseapp.com/03/"
res = requests.get(url)
soup = BeautifulSoup(res.text, "html.parser")
```

```python
lis  = soup.select("ul.lst_pop > li")

popular = []
for li in lis:
    popular.append([li.select_one("a").text, li.select_one("span").text])
print(popular)


lis = soup.select("ul.lst_major > li")

majors = []
for li in lis:
    majors.append([li.select_one("a").text, li.select_one("span").text])
print(majors)
    
 
```

```python

def get_li_info(path):

    lis  = soup.select(path)

    result = []
    
    for li in lis:
        result.append([li.select_one("a").text, li.select_one("span").text])
    
    return result


print(get_li_info("ul.lst_pop > li"))
print(get_li_info("ul.lst_major > li"))
    
```

### 실습 02

- 사이트내에 인기검색종목과 주요해외지수를 각각 클롤링하여 종목명과 상한, 하한 여부를 아래와 같이 리스트로 정리해주세요.

```python
import re

lis  = soup.select("ul.lst_pop > li")
populars = []
for li in lis:
    print(re.search('alt="([가-힣]+)"', str(li.select_one("img"))).group(1))
    
```

```python
lis  = soup.select("ul.lst_pop > li")
populars = []
for li in lis:
    populars.append([li.select_one("a").text, li.attrs['class'][0].replace("up", "상한").replace("dn", "하한")])
print(populars)

lis = soup.select("ul.lst_major > li")
majors = []
for li in lis:
    majors.append([li.select_one("a").text, li.attrs['class'][0].replace("up", "상한").replace("dn", "하한")])
print(majors)
```

### 실습 04

- 분양중인 아파트 정보를 크롤링하여 아래와 같이 딕셔너리 형태로 정리해주세요.

```python
lis = soup.select("ul.sale_list._sale_list > li.sale_item")
apt_infos = []
for li in lis:
    apt_info = {"이름": li.select_one("a").text, 
                "분양가":li.select("dd.txt")[0].select_one("strong").text.replace(",", ""), 
                "유형":li.select("dd.txt")[1].text.split("|")[0].strip(),
                "분양유형":li.select("dd.txt")[1].text.split("|")[1].strip(),
                "세대수":li.select("dd.txt")[2].text.split("|")[0].strip(),
                "평형":li.select("dd.txt")[2].text.split("|")[1].strip()}
    
    apt_infos.append(apt_info)
print(apt_infos)
```

```python
URL = "https://finance.naver.com/sise/sise_quant.nhn"
response = requests.get(URL)
soup = BeautifulSoup(response.text, "html.parser")
tr_list = soup.select("#contentarea > div.box_type_l > table tr")
len(tr_list)
```

###### 2.1.1 품목명과 현재가를 크롤링해주세요.

```python
for tr in tr_list:
    if tr.select_one(".no"):
        td_list = tr.select("td")
        print("{} {} {}".format(
            td_list[0].text, td_list[1].text, td_list[2].text
        ))
```

###### 2.1.2 전일대비 상승한 항목만 품목명, 현재가, 전일비를 크롤링해주세요.

```python
for tr in tr_list:
    if tr.select_one(".no"):
        td_list = tr.select("td")
        if td_list[3].select_one(".red02"):
            print("{} {} {} {}".format(
                td_list[0].text, td_list[1].text, td_list[2].text, td_list[3].text.strip()
            ))

```

### 실습 02 

- 무신사 쇼핑몰의 상의(TOP) 카테고리 첫 페이지의 제품들의 브랜드명, 제품명, 가격을 아래와 같이 크롤링해주세요. (약 10페이지 크롤링)

```python

for i in range(1, 11):
    url = "https://search.musinsa.com/category/001?device=&d_cat_cd=001&brand=&rate=&page_kind=search&list_kind=small&sort=pop&sub_sort=&page={}&display_cnt=90&sale_goods=&ex_soldout=&color=&price1=&price2=&exclusive_yn=&size=&tags=&sale_campaign_yn=&timesale_yn=&q=".format(i)
    
    res = requests.get(url)
    soup = BeautifulSoup(res.text, "html.parser")
    
    article_infos = soup.select("#searchList div.article_info")
    for article_info in article_infos:
        print("브랜드 :", article_info.select_one(".item_title").text.strip())
        print("제품명 :", article_info.select_one(".list_info").select_one("a").attrs['title'])
        print("가격 :", article_info.select_one(".price").text.strip().split()[-1])
        print()
```

### 실습 01

- 디자인 정글 메인페이지 데이터 셀레니움으로 크롤링

```python
%%time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

chromedriver = "./chromedriver"

options = webdriver.ChromeOptions()
options.add_argument("headless")

driver = webdriver.Chrome(chromedriver)
# driver = webdriver.Chrome(chromedriver, chrome_options=options)
driver.get('https://www.jungle.co.kr')

# driver.find_elements_by_css_selector('#more')[0].click()
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
driver.implicitly_wait(2)

title_list = driver.find_elements_by_css_selector(".title")
for i, title in enumerate(title_list):
    print(str(i+1), title.text)
```
