### 실습 01
- HTML 문서내에 ID가 cook 인 태그내의 내용을 출력해주세요.

```python
import requests
from bs4 import BeautifulSoup

URL = "https://scrapying-study.firebaseapp.com/01/"
response = requests.get(URL)
response.status_code
```

```python
soup = BeautifulSoup(response.text, "html.parser")
result = soup.find(id="cook")
print(result.text.strip())
```

### 실습 02
- HTML 문서내에 TABLE 내에 th와 td에 있는 값들을 크롤링해 아래와 같은 딕셔너리 형태를 만들어보세요

```python
ths = soup.find_all("th")
th_list = []
for th in ths:
    th_list.append(th.text)
th_list
```

```python
result = []
tbody = soup.find("tbody")
trs = tbody.find_all("tr")

for tr in trs:
    temp = {}
    tds = tr.find_all("td")
    for i, td in enumerate(tds):
        temp[th_list[i]] = td.text
    result.append(temp)
result
```

```python
result = []
tbody = soup.find("tbody")
trs = tbody.find_all("tr")

for tr in trs:
    temp = []
    tds = tr.find_all("td")
    for  td in tds:
        temp.append(td.text)
    result.append(dict(zip(th_list, temp)))

result
```

### 실습 03
- HTML 문서내에 모든 A 태그에 링크된 페이지에 있는 내용을 읽어 출력해주세요.

```python
anchers = soup.find_all("a")
for ancher in anchers:
    link_url = URL + ancher.attrs['href']
    res = requests.get(link_url)
    html = BeautifulSoup(res.text, 'html.parser')
    print(html.find('p').get_text().strip())
```

# 2. CSS 셀렉터를 활용한 크롤링

### 2.1 BeautifulSoup

```python
import requests
from bs4 import BeautifulSoup

url = "https://scrapying-study.firebaseapp.com/02/"
res = requests.get(url)
soup = BeautifulSoup(res.text, "html.parser")
result = soup.select("#title") #soup.find_all(id="title")
result[0].text.strip()
```

```python
type(result[0])
```

```python
soup.select("body > a:nth-child(4)")
```

```python
soup.select_one("body > a:nth-child(4)") 
```

### 2.2 태그 셀렉터 (HTML 태그를 활용) - p, h1, h2

```python
soup.select_one("title").text
```

```python
result = soup.select(".bold")
print(result)
```

### 2.3 class 셀렉터 (class 속성을 활용) - .content

```python
result = soup.select(".bold.blue")
print(result)
```

### 2.4 속성 셀렉터 (태그내 속성을 활용) - h[href], a[target ="_blank"]

```python
soup.select('a[href]')
```

```python
soup.select("a[target='_blank']")
```

```python
soup.select("a[href$='.com']")
```

### 2.5 후손셀렉터 (해당 태그 내에 포함되는 태그) - div p

```python
soup.select("div#winter p")
```

### 2.6 자식 셀렉터 (해당 태그 바로 안에 포함되는 태그) div > p

```python
soup.select("div#winter > p")
```
