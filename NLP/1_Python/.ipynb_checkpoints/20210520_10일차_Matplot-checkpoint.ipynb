### 1.4 plot 옵션

- color : 선의 색조정 <br>
color = 'red' | red, green, blue 등 색의 이름<br>
color = 'r' | r, g, b, y 등 색이름 약자<br>
color = #ff0000 | rgb 색상코드<br>
<br>
- linestyle : 선의 모양 <br>
linestyle = 'solid' | 실선('-')<br>
linestyle = 'dashed' | 파선('--')<br>
linestyle = 'dashdot' | 1점 쇄선('-.')<br>
linestyle = 'dotted' | 점선 (':')<br>
<br>
- 색과 선의 모양 동시<br>
plt.plot(x, y, ':r')<br>
<br>
이 외의 스타일 관련 : https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
x = np.arange(-4.5, 5, 0.5)

y1 = x**2
y2 = x+30
y3 = x**2 + 10

plt.plot(x, y1, color='red', linestyle='dashed')
plt.plot(x, y2, 'b:')
plt.plot(x, y3, color='#00FF00', linestyle='-.')

plt.show()
```

### 1.5 축의 범위 조정
- x축 : plt.xlim(시작, 끝)
- y축 : plt.ylim(시작, 끝)
- x, y축 : plt.axis([x시작, x끝, y시작, y끝])

```python
x = np.linspace(0, 10, 100)
y = x**2

plt.plot(x, y, '--r')
```

```python
plt.plot(x, y, '--r')
plt.xlim(-20, 30)
plt.show()
```

```python
plt.plot(x, y, '--r')
plt.ylim(-20, 120)
plt.show()
```

```python
plt.plot(x, y, '--r')
plt.axis([-20, 30, -20, 120])
plt.show()
```

### 1.6 label 과 legend

- matplotlib 폰트 설정 <br>
For MacOS : plt.rc('font', family = 'AppleGothic') <br>
For windows : plt.rc('font', family = 'Malgun Gothic')<br>
<br>

- 참고 자료<br>
https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html#matplotlib.pyplot.legend<br>
https://stackoverflow.com/questions/4700614/how-to-put-the-legend-out-of-the-plot<br>

```python
plt.rc('font', family = 'Malgun Gothic')
    
x = np.linspace(0, 10, 100)
y1 = x**2
y2 = x**3

# 선 라벨
plt.plot(x, y1, 'r', label='라인1')
plt.plot(x, y2, 'g', label='라인2')

# x축, y축 라벨
plt.xlabel('x 라벨')
plt.ylabel('y 라벨')

# plt.legend(loc=2) # 선 라벨표 위치 (그래프 내)
plt.legend(bbox_to_anchor=(1.04, 1)) # 선 라벨표 위치 (그래프 밖)
plt.show()
```

### 1.8 subplot & gird

```python
plt.figure(figsize=(5, 5)) # 그래프 전체 크기
grid = plt.GridSpec(2, 3) # gridSpec(행, 열)

plt.subplot(grid[0, 0])
plt.plot([1, 2], color='r')

plt.subplot(grid[0, 1])
plt.plot([1, 2], color='g')

plt.subplot(grid[0, 2])
plt.plot([1, 2], color='b')

plt.subplot(grid[1, :2])
plt.plot([1, 2], color='y')

plt.subplot(grid[1, 2])
plt.plot([1, 2], color='c')

plt.show()
```

### 실습 1 (타이타닉 생존자체크)

- 판다스에서 했던 결과를 가지고 아래와 같은 바 그래프를 그리세요.

```python
import pandas as pd
```

```python
df = pd.read_csv('titanic_train.csv')
df.head()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
def changeAge(x):
    if x // 10 == 0:
        return '유아'
    elif x // 10 == 1:
        return '10대'
    elif x // 10 == 2:
        return '20대'
    elif x // 10 == 3:
        return '30대'
    elif x // 10 == 4:
        return '40대'
    elif x // 10 == 5:
        return '50대'
    elif x // 10 == 6:
        return '60대'
    else:
        return '노인'
    
df['AgeRange'] = df['Age'].apply(changeAge)
```

```python
survived_by_ages = df.groupby('AgeRange')['Survived'].sum()
survived_by_ages
```

```python
survived_by_ages = survived_by_ages.reindex(['유아', '10대', '20대', '30대', '40대', '50대', '60대', '노인'])
survived_by_ages.index
```

```python
plt.bar(survived_by_ages.index, survived_by_ages.values, color='g')
plt.title('나이대별 타이타닉 생존자')
plt.xlabel('나이')
plt.ylabel('생존자')
plt.show()
```

```python
survived_by_sex = df.groupby('Sex')['Survived'].sum()
survived_by_sex
```

```python
plt.bar(survived_by_sex.index, survived_by_sex.values, color='g')
plt.title("성별 타이타닉 생존자")
plt.xlabel('성별')
plt.ylabel('생존자')
plt.show()
```

### 실습 2 (코로나 데이터)

- kr_daily.csv 파일을 읽어 아래와 같은 그래프 두개를 한 화면에 그리세요.

```python
corona = pd.read_csv('./kr_daily.csv')
corona.head()
```

```python
corona.info()
```

```python
def change_date(x):
    x = str(x)
    return x[:4] + '-' + x[4:6] + '-' + x[6:]
```

```python
change_date('20200125')
```

```python
corona['date_text'] = corona['date'].apply(change_date)
corona.head()
```

```python
corona.index = corona['date_text']
corona.tail()
```

```python


plt.subplot(121)
plt.title('확진자')
plt.xticks(rotation=70)
plt.plot(corona['2020-05-18':].index, corona['2020-05-18':]['confirmed'].values, color='red', linestyle='dashed')

plt.subplot(122)
plt.title('사망자')
plt.xticks(rotation=70)
plt.bar(corona['2020-05-18':].index, corona['2020-05-18':]['death'].values, color='red', linestyle='dashed')

plt.show()
```
